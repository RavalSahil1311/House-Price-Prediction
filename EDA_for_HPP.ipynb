```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
my_colors = ['Red' ,'Orange' , 'Yellow' , 'Green' , 'Blue' , 'Indigo' , 'Violet']
```

```python
pd.pandas.set_option('display.max_columns',None)
# pd.pandas.set_option('display.max_rows',None)
```

```python
dataset=pd.read_csv('train.csv')
print(dataset.shape)
```

```python
dataset.head()
```

```python
# Missing Values
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
# dataset[features_with_na]
for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4),'- missing Values')
```

```python
for feature in features_with_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar(color=my_colors)
    plt.title(feature)
    plt.show()
```

```python
print("ID of Houses {}".format(dataset['Id'].count()))
```

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
len(numerical_features)
dataset[numerical_features].head()
```

```python
numerical_features
```

```python
# Temporal Variables(Ex.Datatime Variables)
year_feature=[feature for feature in dataset.columns if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot(color='red')
# plt.bar('YEAR SOLD','SALE PRICE' ,color='red')
dataset.groupby('YrSold')['SalePrice'].mean().plot(color='green')
plt.xlabel('YEAR SOLD')
plt.ylabel('SALE PRICE')
plt.title('YEAR SOLD VS SALE PRICE')
```

```python
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature= [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Feature Total count {}".format(len(discrete_feature)))
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python

for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar(color=my_colors)
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title('{} Vs SalePrice'.format(feature))
    plt.show()
```

```python
continuous_feature=[feature for feature in numerical_features if feature not in  discrete_feature and feature not in year_feature+['Id']]
print('continuous feature Total count {}'.format(len(continuous_feature)))
continuous_feature
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.title('LOG Normal Dis.')
        plt.show()
```

```python
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title('Outliers Of '+feature)
        plt.show()
```

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
categorical_features
```

```python
dataset[categorical_features].head()
```

```python
for feature in categorical_features:
    print('for {}_feature number of unique categories are {}'.format(feature,len(dataset[feature].unique())))
```

```python
for feature in categorical_features:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar(color=my_colors)
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```
